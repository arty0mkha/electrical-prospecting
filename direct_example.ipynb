```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt 
from code import direct_problem as direct
```

```python
plt.plot(m,res0, label=f'r={r}')
plt.plot(m,res1, label=f'r={r*2}')
plt.plot(m,res2, label=f'r={r*4}')
plt.legend()
plt.title('Зависимость интегрируемой функции от r')
```
