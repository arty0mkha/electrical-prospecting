```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from code_folder.visualization_functions import geological_environment
from code_folder.direct_problem import calculate_apparent_resistance
```

```python
rmin,rmax=0.1,10**3
r=np.logspace(np.log10(rmin),np.log10(rmax),25)

param=[]

for i in range(10,100):
    param.append(10+i/10)
    param.append(1)

param.append(50)
method='U'
```

```python
rk_start=[]
for r_i in r:
    rk_start.append(calculate_apparent_resistance(param,method,r_i,num_of_zeros=10*int(1+r_i/200)))
```

```python
plt.loglog(r,rk_start)
```
