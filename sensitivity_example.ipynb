```python
import matplotlib.pyplot as plt
import numpy as np
from code import direct_problem as direct 
from code import visualization_functions as visfunc
from code import sensitivity as sens
```

```python
r=np.logspace(-1,3,50)
param = [100,20,
         50,40,
         20]
param=[87, 2,
       150, 0.3,
       400, 6.5,
       14.5, 42,
       150, 99,
       1200] # из обратной задачи в другом файле
method='U'
delta=5
N=int((len(param)+1)/2)
resistance=[]
for r_i in r:
    resistance.append(direct.calculate_apparent_resistance(param,method,r_i,10*int(1+r_i/200)))
```

```python
fig=plt.figure(figsize=(16,N*8))
for i in range(2*N-1):
    if i%2==0:
        t=f'сопротивлению {int(i/2+1)}го слоя при изменении на {delta} ом'
    else:
        t=f'мощности {int((i+1)/2)}го слоя при изменении на {delta} метров'
    plt.subplot(N,2,i+1)
    plt.semilogx(r,sens.sense_line(r,param,i,delta,method))
    plt.title(f'Чувстительность к {t}')
    plt.xlabel('Полуразнос')
    plt.ylabel('Степень зависимости')
plt.subplot(N,2,2*N)
plt.imshow(visfunc.geological_environment(param))
plt.title('geological_environment')
plt.colorbar(label='Ом',  ticks=param[0::2])
plt.ylabel('Глубина')
```
